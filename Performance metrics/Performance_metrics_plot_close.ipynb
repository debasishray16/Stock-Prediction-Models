```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set Seaborn theme for the plot
sns.set_theme(style="whitegrid")

# Data
epoch = [75, 100, 125, 150]
precision_5 = [0.5714, 0.5714, 0.6897, 1.0]
precision_8 = [0.9855, 0.9784, 0.9855, 0.9927]
precision_10 = [0.9351, 1.0, 0.9828, 0.9713]

# Plot setup
fig, ax = plt.subplots(figsize=(16, 9))
ax.set_facecolor('#ffffff')

# Plot each line with Seaborn
sns.lineplot(x=epoch, y=precision_5, label='5 Year Price', marker="o", color="blue", ax=ax)
sns.lineplot(x=epoch, y=precision_8, label='8 Year Price', marker="o", color="green", ax=ax)
sns.lineplot(x=epoch, y=precision_10, label='10 Year Price', marker="o", color="red", ax=ax)

# Customize plot appearance
plt.grid(True, linestyle='--', color='#BDBDBD')
plt.xticks(ticks=[75, 100, 125, 150])
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Epoch vs Precision')

# Add annotations for each point
for x, y in zip(epoch, np.round(precision_5, 4)):
    plt.text(x, y, f'({x},{y})', ha='right', fontsize=15)

for m, n in zip(epoch, np.round(precision_8, 4)):
    plt.text(m, n, f'({m},{n})', ha='right', fontsize=15)

for p, q in zip(epoch, np.round(precision_10, 4)):
    plt.text(p, q, f'({p},{q})', ha='right',fontsize=15)

plt.legend()
plt.show()
```
