```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(16,9))
fig.patch.set_facecolor('#ffffff')

ax=plt.axes()
ax.set_facecolor('#ffffff')

plt.grid(True,linestyle='--',color='#BDBDBD')
plt.tight_layout()

epoch=[75,100,125,150]
precision_5=[0.5714285714285714 ,0.5714245714285714 , 0.6896551724137931 , 1.0 ]
precision_8=[0.9855072463768116 ,0.9784172661870504 ,0.9855072463768116 ,0.9927007299270073 ]
precision_10=[0.9351351351351351 ,1.0, 0.9827586206896551 ,0.9712643678160919  ]
plt.plot(epoch, precision_5, label='5 Year Price')
plt.plot(epoch, precision_8,label='8 Year Price')
plt.plot(epoch, precision_10,label='10 Year Price')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Epoch vs Precision')

for x,y in zip(epoch, np.round(precision_5,4)):
    plt.text(x, y, f'({x},{y})')

for m,n in zip(epoch, np.round(precision_8,4)):
    plt.text(m, n, f'({m},{n})')

for p,q in zip(epoch, np.round(precision_10,4)):
    plt.text(p, q, f'({p},{q})')

plt.xticks(ticks=[75,100,125,150]) # set the xticks to be 50, 100, 125, 150

plt.show()
plt.style.use('ggplot')
```
