```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(16,9))
fig.patch.set_facecolor('#ffffff')

ax=plt.axes()
ax.set_facecolor('#ffffff')

plt.grid(True,linestyle='--',color='#BDBDBD')
plt.tight_layout()

epoch=[75,100,125,150]
precision_5 = [0.49056603773584906,0.58,0.6666666666666666,0.6]
precision_8 = [1.0,0.9926470588235294,1.0 ,0.9854014598540146 ]
precision_10 = [0.9767441860465116,0.957983193277311,0.9658119658119658,0.98 ]
plt.plot(epoch, precision_5, label='5 Years')
plt.plot(epoch, precision_8,label='8 Years')
plt.plot(epoch, precision_10,label='10 Years')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Epoch vs Precision (Open)')

for x,y in zip(epoch, np.round(precision_5,4)):
    plt.text(x, y, f'({x},{y})')

for m,n in zip(epoch, np.round(precision_8,4)):
    plt.text(m, n, f'({m},{n})')

for p,q in zip(epoch, np.round(precision_10,4)):
    plt.text(p, q, f'({p},{q})')

plt.xticks(ticks=[75,100,125,150]) # set the xticks to be 50, 100, 125, 150

plt.show()
plt.style.use('ggplot')
```
