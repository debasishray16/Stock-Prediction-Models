```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(16,9))
fig.patch.set_facecolor('#ffffff')

ax=plt.axes()
ax.set_facecolor('#ffffff')

plt.grid(True,linestyle='--',color='#BDBDBD')
plt.tight_layout()

epoch=[75,100,125,150]
rsquare_with=[0.9984937478879895,0.9984937478879895,0.9984937478879895,0.9984937478879895  ]
rsquare_without=[0.6154103765940974,0.6142467853634768,0.5501064001507697,0.6621667315880608 ]
plt.plot(epoch, rsquare_with, label='5 Years with XG-Boost')
plt.plot(epoch, rsquare_without,label='5 Years without XG-Boost')


plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Rsquare')
plt.title('Model Efficiency: Epoch vs Rsquare (Open)')

for x,y in zip(epoch, np.round(rsquare_with,4)):
    plt.text(x, y, f'({x},{y})')

for m,n in zip(epoch, np.round(rsquare_without,4)):
    plt.text(m, n, f'({m},{n})')

plt.xticks(ticks=[75,100,125,150]) # set the xticks to be 50, 100, 125, 150

plt.show()
plt.style.use('ggplot')
```
