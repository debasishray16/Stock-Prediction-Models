```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(16,9))
fig.patch.set_facecolor('#42A5F5')

ax=plt.axes()
ax.set_facecolor('#C7E1F4')

plt.grid(True,linestyle='--',color='#BDBDBD')
plt.tight_layout()

epoch=[75,100,125,150]
imaginaryLine=[0,0,0,0]
rsquare_with=[0.9977529443578234,-1.4225250954147786,0.9977486706429847,0.9977486706429847 ]
rsquare_without=[-0.058929774113581024,-0.058929774113581124,0.5365637935701698,0.5131133637123044  ]
plt.plot(epoch, rsquare_with, label='5 Years with XG-Boost')
plt.plot(epoch, imaginaryLine,'#9e9e9e')
plt.plot(epoch, rsquare_without,label='5 Years without XG-Boost')


plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Rsquare')
plt.title('Model Efficiency: Epoch vs Rsquare (Close)')

for x,y in zip(epoch, rsquare_with):
    plt.text(x, y, f'({x},{y})')

for m,n in zip(epoch, rsquare_without):
    plt.text(m, n, f'({m},{n})')

plt.xticks(ticks=[75,100,125,150]) # set the xticks to be 50, 100, 125, 150

plt.show()
plt.style.use('ggplot')
```
